## Read Data

```python
import pandas as pd

accidentData = pd.read_csv("../Data/accident.csv", encoding='cp1252')
accidentData.info()
```

## Categorize Data for the convenient

```python
accidentData['Type of Vehicle'] = accidentData['Type of Vehicle'].replace(to_replace =['Bus or coach (17 or more passenger seats)','Minibus (8 – 16 passenger seats)'],  value ="Bus") 

accidentData['Type of Vehicle'] = accidentData['Type of Vehicle'].replace(to_replace =['Car','Taxi/Private hire car'],  value ="car") 


accidentData['Type of Vehicle'] = accidentData['Type of Vehicle'].replace(to_replace =['M/cycle 50cc and under','Motorcycle over 125cc and up to 500cc','Motorcycle over 50cc and up to 125cc','Motorcycle over 500cc'],  value ="Motarcycle") 

accidentData['Type of Vehicle'] = accidentData['Type of Vehicle'].replace(to_replace =['Pedal cycle','Other Vehicle'],  value ="Pedal cycle") 


accidentData['Type of Vehicle'] = accidentData['Type of Vehicle'].replace(to_replace =['Goods vehicle 3.5 tonnes mgw and under','Goods vehicle over 3.5 tonnes and under 7.5 tonnes mgw','Goods vehicle 7.5 tonnes mgw and over'],  value ="Goods Vehicle") 

accidentData['Type of Vehicle'] = accidentData['Type of Vehicle'].replace(to_replace =['Agricultural vehicle (includes diggers etc.)'], value ="Agri Vehicle") 
```

```python
accidentData['Type of Vehicle'].unique()
```

## Encode The Categorical Features

```python
#Apply label Encoding for the ordinal feaures
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 
accidentData['1st Road Class'] = label_encoder.fit_transform(accidentData['1st Road Class']) 
accidentData['Road Surface']   = label_encoder.fit_transform(accidentData['Road Surface']) 
accidentData['Lighting Conditions'] = label_encoder.fit_transform(accidentData['Lighting Conditions']) 
accidentData['Weather Conditions'] = label_encoder.fit_transform(accidentData['Weather Conditions']) 
accidentData['Casualty Class'] = label_encoder.fit_transform(accidentData['Casualty Class']) 
accidentData['Sex of Casualty'] = label_encoder.fit_transform(accidentData['Sex of Casualty']) 
accidentData['Type of Vehicle'] = label_encoder.fit_transform(accidentData['Type of Vehicle']) 
```
