## Dataset Summary
    
#Total number of features: 15

1) Reference Number - Registered Accident Number
2) Easting    -  Easting and Northing are geographical cartesian co-ordinates for a point. Easting is the eastward measured distance(X axis)    
3) Northing  -    Northing is the northward measured distance(Y axis). easting and northing commonly measured in meters from the axes of some horizondal datum.    
4) Number of Vehicles- Number of vehicles lead to Accident   
5) Accident Date 
6) Time (24hr) - Accident time     
7) 1st Road Class - Types based on road Quality level 
8) Road Surface   - road surface level    
9) Lighting Conditions    - Lighting condition while the accident
10) Weather Conditions    - weather condition while the accident 
11) Casualty Class  - who affected/ lead by the accident 
12) Sex of Casualty    - sex of the person who affected/ lead by the accident 
13) Age of Casualty    - Age of the person who affected/ lead by the accident 
14) Type of Vehicle - Type of the vehicle which affected/ lead by the accident 

#target
15) Casualty Severity  - contains three target attribute which is about the casuality level of the accident

**Total number of samples: 3057 samples
**Total Number of Missing Values: 0
